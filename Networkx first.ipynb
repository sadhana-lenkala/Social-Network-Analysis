```python
#creating custom delete functions
def delete_node():
    G = nx.Graph()
    G.add_edges_from([('A','B'),('B','C'),
                       ('C','A'),('C','D')])
    nx.draw_networkx(G,with_labels=True,
                     node_color='c',node_size=500)
    #remove node
    G.remove_node('B')
    nx.draw_networkx(G,with_labels=True,
                     node_color='r',node_size=500)
```

```python
delete_node()
```

```python
#Deleting specific edges
def delete_edge():
    DE = nx.Graph()
    DE.add_edges_from([('A','B'),('B','C'),
                       ('C','D'),('A','C')])
    nx.draw_networkx(DE,with_labels=True,
                        node_color = 'c', node_size =500)
    #remove edges
    DE.remove_edge('A','B')
    nx.draw_networkx(DE,with_labels=True,
                        node_color = 'r', node_size =500)
```

```python
delete_edge()
```

```python
#custom node size and node color in graph
def custom_NS_NC():
    J = nx.Graph()
    J.add_edges_from([('a','b'),('c','d'),
                      ('a','c'),('b','c')])
    nx.draw_networkx(J)
    plt.show()
    
    #define custom node color
    nodelst = nx.nodes(J)
    colorlst = ['c','y','b','r']
    sizelst = [1000,500,100,2000]
    
    nx.draw_networkx(J,with_labels=True,
            node_color=colorlst,node_size=sizelst,
            nodelist = nodelst)
```

```python
custom_NS_NC()
```

## Bipartite graphs : Two set of nodes and two different set of graphs


```python
from networkx.algorithms import bipartite
B = nx.Graph()
```

```python
cities = ['NYC','LA','LV','VC','Miami','SF','San Diego','MV']
```

```python
people = ['Sadhana','Student']
```

```python
#adding edges of cities and people
B.add_nodes_from(cities, bipartite = 'cities')
B.add_nodes_from(people, bipartite = 'people')

#define edges
edges = [('Sadhana','NYC'),('Student','LA'),('Sadhana','SF')
        ,('Student','MV'),('Student','Miami'),('Sadhana','LV'),
         ('Sadhana','VC'),('Student','San Diego'),]
B.add_edges_from(edges)
```

```python
nx.draw(B,with_labels=True)
```
